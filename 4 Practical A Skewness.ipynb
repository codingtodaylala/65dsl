```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Example: fill missing values in 'math score' with mean
df['math score'] = df['math score'].fillna(df['math score'].mean())

# Example: drop rows with missing values in 'reading score'
df.dropna(subset=['reading score'], inplace=True)


```

```python
df['math score'].skew()
```

```python
df['reading score'].skew()
```

```python
df['writing score'].skew()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Score')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['reading score'], kde=True)
plt.title('Distribution of reading Score')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['writing score'], kde=True)
plt.title('Distribution of writing Score')
plt.show()

```

```python
df['math score power'] = df['math score']**1.5
print("Math score skewness after applying Power transformation: ", df['math score power'].skew())

df['reading score power'] = df['reading score']**1.5
print("Reading score skewness after applying Power transformation: ",df['reading score power'].skew())

df['writing score power'] = df['writing score']**1.5
print("Writing score skewness after applying Power transformation: ",df['writing score power'].skew())



```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math score power'],kde=True)
plt.title("Math after removing skewness")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['reading score power'],kde=True)
plt.title("Read after removing skewness")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['writing score power'],kde=True)
plt.title("Write after removing skewness")
plt.show()
```
